# 2D Translation

```python
import matplotlib.pyplot as plt

# Points of original triangle
triangle = [(1, 1), (3, 1), (2, 4)]

# Translation values
tx, ty = 2, 3

# Translate points
translated = [(x + tx, y + ty) for x, y in triangle]

# Close the triangles
triangle += [triangle[0]]
translated += [translated[0]]

# Plot both triangles
plt.plot(*zip(*triangle), label='Original', color='blue')
plt.plot(*zip(*translated), '--', label='Translated', color='green')

plt.title('Triangle Translation')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()
```

# 2D Scaling

```python
import matplotlib.pyplot as plt

# Original triangle points
triangle = [(1, 1), (3, 1), (2, 4)]

# Scaling factors
sx, sy = 2, 2

# Scale points
scaled = [(x * sx, y * sy) for x, y in triangle]

# Close the triangles
triangle += [triangle[0]]
scaled += [scaled[0]]

# Plot both triangles
plt.plot(*zip(*triangle), label='Original', color='blue')
plt.plot(*zip(*scaled), '--', label='Scaled', color='orange')

plt.title('Triangle Scaling')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()
```

 # 2D Rotation

```python
import matplotlib.pyplot as plt
import math

# Original triangle
triangle = [(1, 1), (3, 1), (2, 4)]

# Rotation angle in degrees
angle = 90
theta = math.radians(angle)

# Rotate points
rotated = [
    (round(x * math.cos(theta) - y * math.sin(theta), 2),
     round(x * math.sin(theta) + y * math.cos(theta), 2))
    for x, y in triangle
]

# Close triangles
triangle += [triangle[0]]
rotated += [rotated[0]]

# Plot both
plt.plot(*zip(*triangle), label='Original', color='blue')
plt.plot(*zip(*rotated), '--', label=f'Rotated ({angle})', color='red')

plt.title('Triangle Rotation')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()
```

# 2D Reflection

```python
import matplotlib.pyplot as plt

def plot_polygon(points, *args, **kwargs):
    x, y = zip(*points)
    plt.plot(x + (x[0],), y + (y[0],), *args, **kwargs)

# Original triangle
triangle = [(3, 4), (6, 4), (3, 6)]

# Plot all triangles
plot_polygon(triangle, label='Original', color='blue')
plot_polygon([(x, -y) for x, y in triangle], '--', label='Reflected over X-axis', color='green')
plot_polygon([(-x, y) for x, y in triangle], ':', label='Reflected over Y-axis', color='red')

# Add labels and formatting
plt.title('Triangle Reflection')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()
```

# 2D shearing

```python
import matplotlib.pyplot as plt

# Helper function to plot a closed polygon
def plot_polygon(points, *args, **kwargs):
    x, y = zip(*points)
    plt.plot(x + (x[0],), y + (y[0],), *args, **kwargs)

# Original triangle
triangle = [(1, 1), (0, 0), (1, 0)]
shear_x, shear_y = 1.0, 1.0  # Shearing factors

# Plot all triangles
plot_polygon(triangle, label='Original', color='blue')
plot_polygon([(x + shear_x * y, y) for x, y in triangle],
             '--', label='Sheared in X', color='purple')
plot_polygon([(x, y + shear_y * x) for x, y in triangle],
             ':', label='Sheared in Y', color='orange')

# Formatting
plt.title('Triangle Shearing')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()
```
